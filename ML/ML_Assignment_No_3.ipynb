#### Problem Statement :- 
    Given a bank customer, build a neural network-based classifier that can determine whether they will leave or not in the next 6 months. Perform following steps: 
       1. Read the dataset. 
       2. Distinguish the feature and target set and divide the data set into training and test sets. 
       3. Normalize the train and test data. 
       4. Initialize and build the model. Identify the points of improvement and implement the same. 
       5. Print the accuracy score and confusion matrix (5 points).
       
Dataset link : https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling 

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(['RowNumber','Surname', 'CustomerId'], axis = 'columns', inplace =True)
df
```

```python
le = LabelEncoder()
df.Gender = le.fit_transform(df.Gender)
df.Geography = le.fit_transform(df.Geography)
df
```

```python
df['Exited'].value_counts()
```

```python
x = df.drop('Exited', axis= 'columns')
y = df['Exited']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state =1)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
model = keras.Sequential([
    keras.layers.Dense(13, input_shape=(10,),activation='relu'),
    # input shape 10 bcoz first layer has inputs for activation
    #relu => max(y, 0) => clamp negative values to 0
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# loss => cost function
# optimizer => algorithm to use to find minima of cost function
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
yp = model.predict(x_test)
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_pred)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
